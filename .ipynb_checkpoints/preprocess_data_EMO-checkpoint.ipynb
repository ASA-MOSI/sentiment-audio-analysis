```python
import os
import re
import glob 
import matplotlib.pyplot as plt
import numpy as np
from os.path import basename
from multiprocessing import Pool
import sys
import numpy
import librosa
import pickle
import acousticFeatures
from typing import Dict, Tuple, List



AUDIO_DATA_PATH = "auditary_emotion_recognition/EMO-DATA/wav"
input = []
output = []
for f in os.listdir(AUDIO_DATA_PATH):
    if not f.startswith(".") and os.path.splitext(f)[1] == ".wav":
        # Get file name
        code = os.path.splitext(f)[0] 
       
        emotion = code[5]
                
        # Get file path
        file_path = "/".join([AUDIO_DATA_PATH, code]) + ".wav"
        
        try:
            features = acousticFeatures.getAllFeatures(file_path)
            print(emotion)
            input.append(features)
            output.append(emotion)
            
        except:
            print("May be file too short.")
        

file_input_handle = open("processed-data/input-EMO.obj", 'wb')
file_output_handle = open("processed-data/output-EMO.obj", 'wb')

pickle.dump(input, file_input_handle)
pickle.dump(output, file_output_handle)
print("Finish")
```
